* Importing the Libaries

```python
import numpy as np 
import pandas as pd
import difflib 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#data Collection and pre-Precossing
```

```python
df=pd.read_csv(r"D:\python\python_dataset\movies.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
#selecting the relvent  fetures for recommentation
```

```python
selected_feture=['genres','keywords','tagline','cast','director']
selected_feture
```

```python
#replacing the null values with null string
```

```python
for feature in selected_feture: 
    df[feature]=df[feature].fillna('')
```

```python
#combing all the 5 selected ffeature 
combined_feature=df['genres']+' '+df['keywords']+' '+df['tagline']+' '+df['cast']+' '+df['director'] 
```

```python
print(combined_feature)
```

```python
#converting the text data to feature vector 

vectorizer = TfidfVectorizer()
```

```python
feature_vectors=vectorizer.fit_transform(combined_feature)
```

```python
print(feature_vectors)
```

<h4>Cosine similarity</h4>

```python
# getting the similarity score using cosine similarity
```

```python
similarity=cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
#getting the movie name from the user 
movie_name=input('enter your favourite movie name:')
```

```python
#creating  a list with all name given in the dataset
```

```python
list_of_all_titles=df['title'].tolist()
print(list_of_all_titles)
```

```python
#Finding the close match for the movie name give by the user 
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles) 
print(find_close_match)
```

```python
close_match=find_close_match[0]
```

```python
print(close_match)
```

```python
#find the index of the movie with title 
index_of_the_movie=df[df.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
#getting a list of similar movie 
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#sorting the movie based on their ssimilarity score 
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

```python
#print the name of similar movies based on the index 
print('movies suggested for you: \n') 
i=1 
for movie in sorted_similar_movies: 
    index=movie[0]
    title_from_index = df[df.index==index]['title'].values[0]
    if(i<25):
        print(i,'-',title_from_index)
        i=i+1
```

<h1 style="text-align:center;">Movie Recommendation</h1>

```python
movie_name=input('enter your favourite movie name:')

list_of_all_titles=df['title'].tolist()

find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles) 

close_match=find_close_match[0]

index_of_the_movie=df[df.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)

print('movies suggested for you: \n') 
i=1 
for movie in sorted_similar_movies: 
    index=movie[0]
    title_from_index = df[df.index==index]['title'].values[0]
    if(i<25):
        print(i,'-',title_from_index)
        i=i+1
```
